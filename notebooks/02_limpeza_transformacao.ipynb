# Importações

```python
import pandas as pd
import numpy as np
```

```python
url = 'https://raw.githubusercontent.com/tmiqueluzzi/car-sales-data-analysis/refs/heads/main/data/raw/Car%20Sales.xlsx%20-%20car_data.csv'
df = pd.read_csv(url)
df.head()
```

# Tratamento

```python
df.columns = (
    df.columns
    .str.strip()
    .str.lower()
    .str.replace(" ", "_")
)
df.columns
```

```python
df['date'] = pd.to_datetime(df['date'])
```

```python
df['year'] = df['date'].dt.year
df['month'] = df['date'].dt.month
df['month_name'] = df['date'].dt.month_name()
```

```python
df.head()
```

# Checagem e exportação

```python
df.info()
df.describe()
```

```python
df.to_csv("car_sales_clean.csv", index=False)
```
